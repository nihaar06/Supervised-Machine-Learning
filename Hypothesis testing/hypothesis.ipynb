```python
sns.histplot(x=taxis.tip,hue=taxis.color,kde=True,palette={'yellow':'yellow','green':'green'})
```

```python
yellow_tip=taxis['tip'][taxis['color']=='yellow']
green_tip=taxis['tip'][taxis['color']=='green']
```

**Hypotheses:**  
- Null Hypothesis (H0): Distribution of tip for yellow=Distribution of tip for green 
- Alternative Hypothesis (H1): Distribution of tip for yellow ≠ Distribution of tip for green 

```python
yellow_tip
```

```python
green_tip
```

```python
print(green_tip.shape)
print(yellow_tip.shape)
```

```python
from scipy import stats

t_stat,p_val=stats.ttest_ind(yellow_tip,green_tip)
print("Independent T-Test statistic:",t_stat)
print("P-Value:",p_val)
```

```python
alpha=0.05
if p_val<alpha:
    print("Reject null hypothesis:The tip for yellow category is different than that of green")
else:
    print("Fail to reject null hypothesis.")
```

**Hypotheses:**

H0 : Distribution of tip is same across all pickup_borough categories

H1 : At least one category has different distribution 

```python
manh_pb=taxis['tip'][taxis['pickup_borough']=='Manhattan']
q_pb=taxis['tip'][taxis['pickup_borough']=='Queens']
brx_pb=taxis['tip'][taxis['pickup_borough']=='Bronx']
br_pb=taxis['tip'][taxis['pickup_borough']=='Brooklyn']

f_stat,p_val=stats.f_oneway(manh_pb,q_pb,brx_pb,br_pb)
print("ANOVA f statistic value:",f_stat)
print("P-value:",p_val)
alpha=0.05
if p_val<alpha:
    print("Reject null hypothesis.")
else:
    print("Fail to reject null hypothesis.")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey_res=pairwise_tukeyhsd(endog=taxis.tip,groups=taxis.pickup_borough,alpha=0.05)
print(tukey_res)
```
